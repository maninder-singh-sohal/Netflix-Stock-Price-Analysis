```python
#plotting volume of stock vs Date
#every 10 consecutive point are avereaged and plotted also

plt.figure(figsize=(20,10))
sns.lineplot(x=df.index,y=df['Volume'],label='Volume')
sns.lineplot(x=df.index,y=df['Volume'].rolling(window=10).mean(),label='Averaged volume')
plt.title('Volume of stock Vs. time')
```

```python
#plotting high,open and close stock price Vs. Date


plt.figure(figsize=(20,10))
sns.lineplot(y=df['High'],x=df.index,label='High')
sns.lineplot(y=df['Open'],x=df.index,label='Open')
sns.lineplot(y=df['Close'],x=df.index,label='Close')
plt.ylabel("stock Price")
plt.title("High, Open, Close stock price daywise ")
```

```python
#creating three dataframes one with index of month days and taken mean for the corresponding records
# second with index of months and taken mean for the corresponding records
# third with index of years and taken mean for the corresponding records

df_day=df.groupby(df.index.day).mean()
df_month=df.groupby(df.index.month).mean()
df_year=df.groupby(df.index.year).mean()
```

```python
# plot 1 : plotting High, Close, Low stock price averaged on same days of every month
# plot 2 : plotting High, Close, Low stock price averaged on same months of every year
# plot 3 : plotting High, Close, Low stock price averaged on year

fig,axes=plt.subplots(nrows=3,ncols=1,figsize=(20,15))
axes[0].plot(df_day.index,df_day[['High','Close','Low']],label=['High','Close','Low'])
axes[1].plot(df_month.index,df_month[['High','Close','Low']],label=['High','Close','Low'])
axes[2].plot(df_year.index,df_year[['High','Close','Low']],label=['High','Close','Low'])
axes[0].set_title('High, Open, Close stock price day wise')
axes[1].set_title('High, Open, Close stock price month wise')
axes[2].set_title('High, Open, Close stock price year wise')
axes[0].legend()
axes[1].legend()
axes[2].legend()
```

```python
#did same plotting for volume of the stock


fig,axes=plt.subplots(nrows=3,ncols=1,figsize=(20,15))
axes[0].plot(df_day.index,df_day['Volume'],label='Volume')
axes[1].plot(df_month.index,df_month['Volume'],label='Volume')
axes[2].plot(df_year.index,df_year['Volume'],label='Volume')
axes[0].set_title('Volume day wise')
axes[1].set_title('Volume month wise')
axes[2].set_title('Volume year wise')
axes[0].legend()
axes[1].legend()
axes[2].legend()
```

Takeaways:

LINEAR REGRESSION

```python
#trying to predict high price of the stock based on open,close and low price of the stock

X = df.drop(['High','Adj Close','Volume'],axis=1)
y = df["High"]
print("Shape Of X", X.shape)
print("Shape Of y", y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=101)
```

```python
lm= LinearRegression()
lm.fit(X_train, y_train)
```

```python
#intercept of our regression line

print(lm.intercept_)
```

```python
#coefficients of our regression line

lm.coef_
```

```python
#creating a dataframe of our coefficients

cdf=pd.DataFrame(lm.coef_,X.columns,columns=['Coef'])
cdf
```

our linear regression equation:
High = 0.025326197298412012 - 0.437289*(Low) + 0.719891*(Open) + 0.725194*(Close)

```python
#predicting the high prices of the stock

predictions=lm.predict(X_test)
predictions
```

```python
#actual high price values

y_test
```

```python
#getting accuracy score 0f 99%

lm.score(X_test,y_test)
```

```python
dff=pd.DataFrame({'X_test':list(y_test),'predictions':list(predictions)})
```

```python
dff.head()
```

```python
#plotting predicted vs actual values of high with regression line

sns.lmplot(data=dff,x='X_test',y='predictions',line_kws={"color": "red"})
```
